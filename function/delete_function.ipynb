## 외국인 제외(한글 없는 포스트 id 제외)

```python
os.chdir('D:\Scrapy\insta_crawling')
df = pd.read_csv('jl_region.csv')
```

```python
id_lst = list(set(df['insta_id']))
len(id_lst)
```

```python
def del_foreign(df):
    id_lst = list(set(df['insta_id']))
    for user_id in id_lst:
        count = 0
        for y in list(df[df['insta_id']==user_id]['content']):
            if type(y) != str:
                pass
            elif re.search('[가-힣]', y):
                count += 1
        if count == 0:
            idx = df[df['insta_id']==user_id].index
            df.drop(idx, inplace=True)
    return df
```

```python
new_df = del_foreign(df)
```

```python
id_lst = list(set(new_df['insta_id']))
len(id_lst)
```

## 팔로워 1000명 이상, 좋아요 1000개 이상 id 제외 

```python
over1000_df = pd.read_csv(r'D:\Scrapy\insta_crawling\jl_over1000.csv')
```

```python
def del_over1000(df, over1000_df):
    # 팔로워 1000명 이상 id 제외
    for over1000_id in list(over1000_df['data']):
        idx = df[df['inner_id']==str(over1000_id)].index
        df.drop(idx, inplace=True)
        
    id_lst = list(set(df['insta_id']))
    print(len(id_lst))
        
    # 좋아요 1000개 이상 게시물 있는 id 제외
    df['like_count'] = df['like_count'].apply(int)
    id_lst = list(set(df['insta_id']))
    for user_id in id_lst:
        count = 0
        for y in list(df[df['insta_id']==user_id]['like_count']):
            if y > 1000:
                count += 1
        if count >= 1:
            idx = df[df['insta_id']==user_id].index
            df.drop(idx, inplace=True)
    id_lst = list(set(df['insta_id']))
    print(len(id_lst))
    return df
```

```python
new_new_df = del_over1000(new_df, over1000_df)
```

```python
id_lst = list(set(new_new_df['insta_id']))
len(id_lst)
```

```python
new_new_df.to_csv('jl_delete.csv', index=False)
```
